# Skill Importance 2025

The following dataset was manually scraped from the [World Economic Forum Future of Jobs 2025 Report](https://reports.weforum.org/docs/WEF_Future_of_Jobs_Report_2025.pdf).

The data in this notebook is found on page 92 of the report in Table 5.7. 

The manually scraped version of the table is located in ```data/WorldEconomicForum/WEF_SkillImportance2025.tsv``` within this repository

## Loading the dataset

````python
import pandas as pd
````

````python
skills_df = pd.read_csv('../../data/WorldEconomicForum/WEF_SkillImportance2025.tsv', sep='\t')

skills_df.set_index('industry_type', inplace=True)

skills_df.head()
````

The first column of the dataset is the industry cluster (as defined by the World Economic Forum). Each following column name is a skill identified by the WEF and the value for each column reflects the percentage of companies within an industry cluster that responded that this skill is will be important in the year 2030.

## Simple insights from the dataset

Taking the mean of each column gives us the skills that across all industries are seen to be the most valuable going into 2030.

According to the report, the top 3 most important skills on average no matter the industry are:

1) Analytical Thinking
2) Resiliance, Flexibility, and Agility
3) Creative Thinking

````python
skills_df.mean(axis=0).sort_values(ascending=False)
````
